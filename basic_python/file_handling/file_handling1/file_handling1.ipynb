```python
import csv
# the file object treats the CSV file as a simpple text file with variuos columns seperated by commas
file_obj = open("year2017.csv", 'r')
file_data = csv.reader(file_obj)
print(type(file_data))
c = 0
for i in file_data:
    print(i)
    if (c == 5):
        break
    c += 1
c = 0
for j in file_data:
    print(j)
    if (c == 5):
        break
    c += 1
file_obj.close()
'''
the csv.reader(file_obj) object returns a User defined container similar to Standard containers, to iterate over it an iterator
is used which in the above code is 'i'.
each element of the reader object is a list of all elements in a given row. each element of the list being the value
between 2 delimiters(here it is a comma) in the CSV file.
in every iteration the iterator using the reader object reads one row of the CSV file and returns after proccesing it,
then the iterator moves to the next line.
if all lines are exhausted then the reader obejct must be reset to reread the entire file.
'''
```

```python
import csv
with open("year2017.csv", 'r') as file_obj:
    file_data = csv.reader(file_obj)
    file_list = list(file_data)
# file_list is a 2-D list with each element as one row of the file 'year2017.csv'
for i in range(5):
    print(file_list[i])
```

```python
# now we want to find the total number of people who were killed a/c to year2017.csv file
# we can do this by iterating over the 'killed' column of the csv file

import csv

with open('year2017.csv', 'r') as file_obj:
    file_reader = csv.reader(file_obj, delimiter=',', skipinitialspace=True)
    # skip initial space argument formats the elements of each row of the file by truncation initial spaces from them
    file_list = list(file_reader)
killed = []
sum=0
# slicing the list from the first element(the header row)
for i in file_list[1:100]:
    if i[9] == '':
        continue
    # 'i' is a list and its 9th index represents the killed column
    killed.append(float(i[9]))
    sum = sum+float(i[9])

print(killed)
print("total killed: ", sum)
# now the the list consists of strings which need to be converted to integers to perform operations on them
'''for this the list first needs to be cleaned i.e. strings with extra spaces need to be formatted and and blank strings
(missing data points) must be ignored'''
```

```python
import csv 
with open('year2017.csv', 'r') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True)
    '''
    the file_data is a container object similar to standard template containers whose each element is a dictionary,
    the dictionary keys are the corresponding header names and values are the values for that row fpor that header column
    '''
    c=0
    for i in file_data:# 'i' is a dictionary
        if(c==10):
            break
        print(i)
        print(i['Country'])
        c=c+1
```

```python
import csv 
with open('year2017.csv', 'r') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True, fieldnames=('a','b','c'))
    '''
    the feildnames parameter assigns custom key names which act as custom headers
    columns to which key names have not been assigned in the feildnames parameter are stored togehter in a list 
    which acts as value corresponding to a key 'none'
    '''
    c=0
    for i in file_data:
        if(c==10):
            break
        print(i)
        c=c+1
```
