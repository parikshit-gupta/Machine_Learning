# loading boston dataset from SciKit learn library

```python
from sklearn import datasets
import pandas as pd
```

```python
housing=datasets.fetch_california_housing()
housing
```

the various attributes of relevenece of the housing object are:
1. data
2. target
3. DESCR
4. feature_names

```python
x=housing.data
target=housing.target
y=housing.feature_names
z=housing.DESCR
```

```python
housing_df=pd.DataFrame(x)
housing_df.columns=y
print(housing_df.head(10))
print(housing_df.shape)
```

```python
print(z)
```

```python
# The train_test_split function helps in splitting the given dataset for training our ML algorithm
from sklearn import model_selection
print(model_selection.train_test_split(x,target))
x_train, x_test, target_train, target_test=model_selection.train_test_split(x,target)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# creating algorithm object
alg1=LinearRegression()
alg1.fit(x_train, target_train)
target_predicted=alg1.predict(x_test)
print(target_predicted)
""" 
    the parameters of the line predicted by the linear regression algorithm
"""
print(alg1.coef_)#the coefficients of the input parameter terms
print(alg1.intercept_) #the intercept of the linear function predicted by LIN reg algo
```

```python
import matplotlib.pyplot as plt

plt.plot([0,1,5],[0,1,5], 'r-')#y=x line to comare predicted and actual outputs
plt.axis([0,5,0,5])
plt.scatter(target_predicted, target_test, s=.1)
plt.title('first algorithm plot')
plt.ylabel('Test_data')
plt.xlabel('predicted_data')
plt.show()
'''
the plot shows that the linear regression algo is working well enough as most of the data points lie near the line y=x
although there are a few stray points, this can get better with more sophisticated algorithms.
'''
```

```python
alg1.score(x_train, target_train)
#calculates the coeff of determination for based on the target values it predicts and the given correct target vales
```
