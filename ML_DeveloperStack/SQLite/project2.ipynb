# SQLite project

```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
db= sqlite3.connect("IMDB.sqlite")
data_frame=pd.read_sql_query("select * from IMDB", db)
data_frame.head(5)
```

```python
voted_array=(np.array(list(data_frame['TotalVotes'])))
index=np.argmax(voted_array)
print("most voted movie is: ", data_frame['Title'].iloc[index], " with rating ", data_frame['Rating'].iloc[index])
```
