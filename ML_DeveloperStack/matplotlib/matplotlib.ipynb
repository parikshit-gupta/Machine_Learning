```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[2,4,6]

x2=[1,5]
y2=[10,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
plt.plot(x,y,'b-')
plt.plot(x,y,'go')
plt.plot(x2,y2,'r--')
plt.plot(x2,y2,'bo')
plt.plot([1,3],[6,5],'y^')
plt.plot([1,3],[6,5],'g-.')
plt.show()
```

```python
#plotting x^3 using matplotlib
import numpy as np
arr=np.arange(-20,20,1)
cube_arr=np.arange(-20,20,1)**3
plt.plot(arr,cube_arr,'ro')
plt.plot(arr,cube_arr,'b-')
```

```python
arr=[i**3 for i in range(1,10,1)]
print(arr)
```

```python
#plotting the graph of sinx and cosx using matplotlib
x1=np.arange(-10,10,.1)
y1=np.sin(x1)
y3=np.cos(x1)
plt.plot(x1,y1,color='red', linewidth=1, label='sin(x)')
plt.plot(x1,y3, color='black', linewidth=1, label='cos(x)')
plt.ylabel('y')
plt.xlabel('x')
plt.title('sin(x) and cos(x)')
plt.legend(loc='upper right')
plt.grid()
plt.axis([-np.pi*3,np.pi*3,-1,1])
```

```python
#plotting multiple plots side by side
x=np.arange(0,20,.1, dtype=float)
Y=np.c_[x, x**2, x**3]
label=['x', 'x**2', 'x**3']
fig, ax=plt.subplots(1,3,figsize=(20,5), sharey=True)#sharey parameter shares the y axis with all the graphs

#ax is a plot object which is used to refer to various plots
for i in range(len(ax)):
    ax[i].scatter(x, Y[:,i], color='b')
    ax[i].set_xlabel(label[i]) 
ax[0].set_ylabel('y')
plt.axis([0,20,0,20])
plt.show()
```
