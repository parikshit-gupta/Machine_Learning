```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#bubble chart
year=np.arange(2010,2024, 1)
avg_sal=np.array([((i+year[i]/2)*100) for i in range(0,14,1)])
num_emp=np.array([(i*i+10) for i in range(0,14,1)])
print(year)
print(avg_sal)
print(num_emp)
```

```python
plt.scatter(year, avg_sal, s=num_emp , c=num_emp)
plt.show()
```

```python
#piechart    
size=[3,4,6,2]
labels=['A', 'B', 'C', 'D']
colors=['red', 'blue', 'green', 'yellow']
plt.pie(size, labels=labels, colors=colors, autopct='%.1f')
plt.show()
```

```python
def f1(size):
    def f2(pct):
        total=sum(size)
        val=int(round(pct*total/100.00))
        return '{p:.2f}% {v:d}'.format(p=pct,v=val)
    return f2
        
size=[3,4,6,2]
labels=['A', 'B', 'C', 'D']
colors=['red', 'blue', 'green', 'yellow']
explode=[.05,.02,0,0]
plt.pie(size, colors=colors, labels=labels, autopct=f1(size), explode=explode)
plt.axis('equal')
plt.show()
```

```python
#string formating using .format() method
print('The valueof pi is: %10.5f' %3.141592)
print('The valueof pi is: {1:1.1f}'.format(3.141592,1.417))
```

```python
#plotting histograms
b=[1,3,4,1,2,4,5,6,1,7,22,10,19,17,14,5,2,1,11]
plt.hist(b, edgecolor='black')
plt.show()
```

```python
plt.hist(b, bins=30, edgecolor='black')
plt.xticks([i for i in range(22)])
plt.show()
```

```python
plt.hist(b, bins=[1,3,6,9,15,17,22], edgecolor='black')
plt.xticks([i for i in range(22)])
plt.show()
```
