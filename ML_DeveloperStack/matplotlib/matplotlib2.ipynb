## Functions learnt during algorithm implementations

    - ax.imshow(numpy_array)
- Renders the numpy_array given into an image along the matplotlib plot's axes.

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# PIL: Python Image Library
image=Image.open("D:\Development\data_science\ML_algorithms\coursera-2\Week-1\Binary_DigitRecognition_ANN_Tensorflow\sab moh maya hai_edited.jpg")
image
```

```python
#converting the image to a numpy array
#the numpy array obtained has 3 dimensions each pixel is a combination of R, G, B values corresponding to that pixel
#the first two dimesions represent the size of the image which is 244x243 pixels

image_np=np.array(image)
image_np.shape
```

```python
fig, ax=plt.subplots() # ax represents a single subplot
ax.imshow(image) #renders the array as an image
```

```python
fig1, ax1=plt.subplots(ncols=3, nrows=2) # now ax1 is a 2D numpy array of matplotlib.axes objects

for i in ax1.flat:
    i.imshow(image)
plt.show()
```
