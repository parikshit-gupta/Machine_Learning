## 'Graphviz'
- Library to visualise graphs.
- The Graphviz library was initially written in the DOT language which was speacially developed for this project.
- Inorder to use Graphviz through python code we need to install a simple python interface module called as 'graphviz'(note the capitalisation of g).
- We will use this library to visualise computational graphs while implementing them from scratch

```python
import graphviz
```

#### graphviz supports 2 types of graphs:
- directed graphs (graphviz.Digraph())
- undirected graphs (graphviz.graph())

```python
dot = graphviz.Digraph(name="Graph1", comment='The Round Table', graph_attr={"rankdir":"LR"})
# the graph_attr attribute decides how to display the graph
# in this case the graph will be displayed from left to right (in order of occurence of nodes)
```

#### constructing the graph:
#### 1) Node
        - graph_obj.node(name="", label="", shape="")
- the name specifed here is used everywhere else to specify the given node
#### 2) edges
        - graph_obj.edge(tail_node, head node, constraint= bool_val)
- the constraint attribute is used to determine if an edge should affect the positioning of the rest of the graph.

```python
dot.node(name='A', label='King | Arthur', shape="record")
dot.node('B', 'Sir Bedevere the Wise')
dot.node('L', 'Sir Lancelot the Brave')
dot.edges(['AB', 'AL'])
dot.edge('B', 'L', constriant='False')
```

```python
print(dot.source)
#prints the DOT source code corresponding to the graph given
```
