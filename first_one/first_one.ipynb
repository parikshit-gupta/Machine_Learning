```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib
```

```python
t=np.arange(0.0,2.0,.01)
s=1+np.sin(2*np.pi*t)
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel="time(s)", ylabel="potty(kgs)", title="potty vs time")
ax.grid()
```
